```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import math

xs = [i / 20 for i in range(0, 130)]
ys = [math.sin(x) for x in xs]

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(xs, ys)
ax.set_title('sine wave — inline plot probe')
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
fig
```
